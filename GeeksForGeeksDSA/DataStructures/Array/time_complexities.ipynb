```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of values for n
# start | end | number of samples to generate
n = np.linspace(1, 100, 4)

# Define the functions
array = n
linked_list = n
log_n = np.log(n)
n_squared = n**2
n_log_n = n * np.log(n)
n_cubed = n**3
two_to_the_n = 2**n
n_factorial = [np.math.factorial(int(x)) for x in n]
n_times_n_plus_1_over_2 = n*(n+1)/2
two_n_minus_1 = 2*n - 1
log_2_n = np.log2(n)
log_3_n = np.log(n) / np.log(3)  # Change of base formula

# Create the plot
plt.figure(figsize=(10, 6))

plt.plot(n, array, label='Array (O(n))')
plt.plot(n, linked_list, label='Linked List (O(n))')
plt.plot(n, log_n, label='Logarithmic (O(log n))')
plt.plot(n, n_squared, label='Quadratic (O(n^2))')
plt.plot(n, n_log_n, label='Linearithmic (O(n log n))')
plt.plot(n, n_cubed, label='Cubic (O(n^3))')
plt.plot(n, two_to_the_n, label='Exponential (O(2^n))')
# plt.plot(n, n_factorial, label='Factorial (O(n!))')
plt.plot(n, n_times_n_plus_1_over_2, label='n*(n+1)/2')
plt.plot(n, two_n_minus_1, label='2n - 1')
plt.plot(n, log_2_n, label='Logarithmic base 2 (O(log_2 n))')
plt.plot(n, log_3_n, label='Logarithmic base 3 (O(log_3 n))')

plt.xlabel('n')
plt.ylabel('Operations')
plt.title('Time Complexities')
plt.legend()
plt.grid(True)

plt.show()
```
